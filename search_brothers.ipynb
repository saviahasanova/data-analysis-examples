```python
import pandas as pd
import numpy as np
import re
```

```python
df_minjust = pd.read_csv('data/Minjust2018.csv', 
                 header=None, 
                 usecols=[4, 27, 28, 33])
df_minjust.fillna('', inplace=True)
df_minjust.rename(columns={4: 'company_name', 
                           27: 'director', 
                           28: 'activity', 
                           33: 'founders'},
                  inplace = True)
```

```python
df_minjust
```

```python
df_decl = pd.read_csv('data/declaration.csv', 
                      header=None, 
                      usecols=[1, 2, 3, 4, 5])
df_decl.fillna('', inplace=True)
df_decl.rename(columns={1: 'surname', 
                        2: 'name', 
                        3: 'patronymic', 
                        4: 'workplace',
                        5: 'position'},
                  inplace = True)
```

```python
df_decl
```

```python
def search_df(df, column, pattern):
    return df.loc[df[column].str.contains(pattern)]
    
def brother_regex(surname, patronymic):
    pattern = re.compile(remove_a(surname) +
                         r'\s+\w+\s+' + # First name
                         remove_vna(patronymic) + # Patronymic
                         r'|' + 
                         r'\w+\s+' + # First name
                         remove_vna(patronymic) + r'\w+\s+' + # Patronymic
                         remove_a(surname),
                         re.IGNORECASE)
    return pattern

def remove_a(surname):
    if surname[-1:] == 'а':
        surname = surname[:-1]
    return surname

def remove_vna(patronymic):
    if patronymic[-3:] =='вна':
            patronymic = patronymic[:-3]
    return patronymic
```

```python
columns = ['company_name', 'director', 'activity', 'founders', 'surname', 'name', 'patronymic', 'workplace', 'position']
result = pd.DataFrame(columns=columns)
```

```python
for index, row in df_decl.loc[1349:1419].iterrows():
    search = search_df(df_minjust, 'founders', brother_regex(row['surname'], row['patronymic'])).copy()
    for column in df_decl.columns:
        search[column] = row[column]
    result = result.append(search, ignore_index=True)
```
